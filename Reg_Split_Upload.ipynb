```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
crime83 = pd.read_csv("83crime.csv")
crime83.rename(columns={"2012 [YR2012]":"Crime" }, inplace=True)  # remove spaces
crime83 = crime83.head(83)
```

```python
education83 = pd.read_csv("83educ.csv")
education83.rename(columns={"2012 [YR2012]":"Education" }, inplace=True)  # remove spaces
education83 = education83.head(83)
```

```python
unemployment83 = pd.read_csv("83unemp.csv")
unemployment83.rename(columns={"2012 [YR2012]":"Unemployment" }, inplace=True)  # remove spaces
unemployment83 = unemployment83.head(83)
gdp83 = pd.read_csv("83gdp.csv")
gdp83.rename(columns={"2012 [YR2012]":"GDP" }, inplace=True)  # remove spaces
gdp83 = gdp83.head(83)
```

```python
data = education83.loc[:,"Education":"Education"]
data["Unemployment"] = unemployment83.loc[:,"Unemployment":"Unemployment"]
data["GDP"] = gdp83.loc[:,"GDP":"GDP"]
data["Crime"] = crime83.loc[:,"Crime":"Crime"]
data
```

```python
# visualize the relationship between the features and the response using scatterplots
fig, axs = plt.subplots(1, 3, sharey=True)
trainData.plot(kind='scatter', x='Education', y='Crime', ax=axs[0], figsize=(16, 8))
trainData.plot(kind='scatter', x='Unemployment', y='Crime', ax=axs[1])
trainData.plot(kind='scatter', x='GDP', y='Crime', ax=axs[2])
```

```python
# this is the standard import if you're using "formula notation" (similar to R)
import statsmodels.formula.api as smf

# create a fitted model in one line
lm = smf.ols(formula='Crime ~ Education', data=trainData).fit()

# print the coefficients
lm.params
```

```python
# create a DataFrame with the minimum and maximum values of TV
X_new = pd.DataFrame({'Education': [trainData.Education.min(), trainData.Education.max()]})
X_new.head()
```

```python
# make predictions for those x values and store them
preds = lm.predict(X_new)
preds
```

```python
# first, plot the observed data
trainData.plot(kind='scatter', x='Education', y='Crime')

# then, plot the least squares line
plt.plot(X_new, preds, c='red', linewidth=3)
```

```python
# print the confidence intervals for the model coefficients
lm.conf_int()
```

```python
# print the p-values for the model coefficients
lm.pvalues
```

```python
# print the R-squared value for the model
lm.rsquared
```

```python
#MULTIPLE REGRESSION
# create a fitted model with all three features
lm = smf.ols(formula='Crime ~ Education + Unemployment + GDP', data=trainData).fit()

# print the coefficients
lm.params
```

```python
# print a summary of the fitted model
lm.summary()
```

```python
# only include Education in the model. We can only reject it.
lm = smf.ols(formula='Crime ~ Education', data=trainData).fit()
lm.rsquared
```

```python
# add Unemployment and GDP to the model (which we believe has no association with Crime)
lm = smf.ols(formula='Crime ~ Education + Unemployment + GDP', data=trainData).fit()
lm.rsquared
```

```python
# create X and y
feature_cols = ['Education', 'Unemployment', 'GDP']
X = trainData[feature_cols]
y = trainData.Crime

# follow the usual sklearn pattern: import, instantiate, fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X, y)

# print intercept and coefficients
lm.intercept_, lm.coef_
```

```python
# pair the feature names with the coefficients
zp = zip(feature_cols, lm.coef_)
dict(zp)
```

```python
# predict for a new observation
lm.predict(t_features)
```

```python
# calculate the R-squared
lm.score(X, y)
```

```python
# calculate the R-squared
lm.score(t_features, t_targetarray)
```
