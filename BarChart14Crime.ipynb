```python
import pandas as pd
import matplotlib as plt
data = pd.read_csv('14countries.csv',index_col=['Country Code'])
```

```python
data
```

```python
country1 = data.iloc[0] # The index of the country you have chosen, E.g. Spain
country2 = data.iloc[5] # Index 1, Japan.
country3 = data.iloc[1]
country4 = data.iloc[12]
clist = [country1,country2,country3,country4]
```

```python
clist
```

```python
for t in range (0,4):
    clist[t] = clist[t][13:22]
```

```python
print(clist)
```

```python
vc1 = []
vc2 = []
vc3 = []
vc4 = []
vc = [vc1,vc2,vc3,vc4]
for g in range(0,4):
    for z in range (0,9):
        num = float(clist[g].get_value(z))
        vc[g].append(num)
```

```python
vc
```

```python
namelist = data.columns.values
namelist = namelist[13:22]
print(namelist)
```

```python
for p in range (0,9):
    namelist[p] = namelist[p][0:4]
```

```python
print(namelist)
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# data to plot
n_groups = 9 # number of years
fsize = 14
# create plot
fig, ax = plt.subplots(figsize=(15, 7))
index = np.arange(n_groups)
bar_width = 0.35 / 2
opacity = 0.8
 
rects1 = plt.bar(index, vc[0], bar_width,
                 alpha=opacity,
                 color='mediumspringgreen',
                 label='Spain')
 
rects2 = plt.bar(index + bar_width, vc[1], bar_width,
                 alpha=opacity,
                 color='hotpink',
                 label='Australia')

rects3 = plt.bar(index + bar_width*2, vc[2], bar_width,
                 alpha=opacity,
                 color='dodgerblue',
                 label='Japan')
 
rects4 = plt.bar(index + bar_width*3, vc[3], bar_width,
                 alpha=opacity,
                 color='sandybrown',
                 label='Morocco')
 
plt.xlabel('years', fontsize = fsize)
plt.ylabel('Intentional homicides (per 100,000 people)', fontsize = fsize - 2)
plt.title('Crime Rates of Countries', fontsize = fsize)
plt.xticks(index + bar_width*3/2, namelist, fontsize = fsize)
plt.yticks(fontsize = fsize)
plt.legend()
 
plt.tight_layout()
plt.show()
fig.savefig("NEWcrime14.png")
```
