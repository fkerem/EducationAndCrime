```python
import pandas as pd
import matplotlib as plt
data = pd.read_csv('educ14count.csv')
```

```python
Country1 = data[["Country","Time","Value"]][(data.LOCATION == 'ESP') & (data.Time > 2003) & (data.Time < 2013)].Value
Country2 = data[["Country","Time","Value"]][(data.LOCATION == 'MAR') & (data.Time > 2003) & (data.Time < 2013)].Value
Country3 = data[["Country","Time","Value"]][(data.LOCATION == 'JPN') & (data.Time > 2003) & (data.Time < 2013)].Value
Country4 = data[["Country","Time","Value"]][(data.LOCATION == 'AUS') & (data.Time > 2003) & (data.Time < 2013)].Value

Country1 = Country1.tolist()
Country2 = Country2.tolist()
Country3 = Country3.tolist()
Country4 = Country4.tolist()

vc = [Country1,Country2,Country3,Country4]
```

```python
print(vc)
```

```python
namelist = data[["Time","Value"]][(data.LOCATION == 'ESP') & (data.Time > 2003) & (data.Time < 2013)].Time
namelist = namelist.tolist()
namelist
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# data to plot
n_groups = 9 #yıl sayısı
fsize = 14
# create plot
fig, ax = plt.subplots(figsize=(15, 7))
index = np.arange(n_groups)
bar_width = 0.35 / 2
opacity = 0.8
 
rects1 = plt.bar(index, vc[0], bar_width,
                 alpha=opacity,
                 color='mediumspringgreen',
                 label='Spain')
 
rects2 = plt.bar(index + bar_width, vc[3], bar_width,
                 alpha=opacity,
                 color='hotpink',
                 label='Australia')

rects3 = plt.bar(index + bar_width*2, vc[2], bar_width,
                 alpha=opacity,
                 color='dodgerblue',
                 label='Japan')
 
rects4 = plt.bar(index + bar_width*3, vc[1], bar_width,
                 alpha=opacity,
                 color='sandybrown',
                 label='Morocco')
 
plt.xlabel('years', fontsize = fsize)
plt.ylabel('Gross enrollment ratio, tertiary, both sexes (%)', fontsize = fsize - 2)
plt.title('Educational Levels of Countries (2004 - 2012)', fontsize = fsize)
plt.xticks(index + bar_width*3/2, namelist, fontsize = fsize)
plt.yticks(fontsize = fsize)
plt.legend()
 
plt.tight_layout()
plt.show()
fig.savefig("14ccbareduc.png")
```
